```sql
-- Seleccionar el nombre, albumId y el compositor de los tracks y ademas el genero del mismo. HINT: La informacion esta repartida entre la tabla ”Track” y la tabla ”Genre”
SELECT t.Name, t.AlbumId, t.Composer, g.Name as 'Genre'
FROM Track t, Genre g
WHERE t.GenreId = g.GenreId
```

```sql
-- 6. A la consulta anterior, agregar el MediaType del track.
SELECT t.Name, t.AlbumId, t.Composer, g.Name as 'Genre', m.Name as 'Media Type'
FROM Track t, Genre g, MediaType m
WHERE t.GenreId = g.GenreId AND t.MediaTypeId = m.MediaTypeId
```

```sql
-- 7. Listar la cantidad de tracks que tiene cada genero y el nombre del genero.
SELECT Genre, COUNT(Genre) AS 'Count' FROM
(SELECT t.Name, t.AlbumId, t.Composer, g.Name as 'Genre'
FROM Track t, Genre g
WHERE t.GenreId = g.GenreId) AS Tablita GROUP BY Genre

SELECT g.Name, COUNT(*) as 'Count'
FROM Track t
JOIN Genre g on t.GenreId = g.GenreId
GROUP By G.Name 
```

```sql
-- 8. Ordenar el resultado de la consulta anterior, de manera descendente por cantidad de tracks

SELECT Genre, COUNT(Genre) AS 'Count' FROM
(SELECT t.Name, t.AlbumId, t.Composer, g.Name as 'Genre'
FROM Track t, Genre g
WHERE t.GenreId = g.GenreId) AS Tablita 
GROUP BY Genre 
ORDER BY [Count] DESC

SELECT g.Name, COUNT(*) as 'Count'
FROM Track t
JOIN Genre g on t.GenreId = g.GenreId
GROUP By G.Name 
ORDER BY [Count] DESC
```

```sql
--9. Obtener los artistas que no tienen albumes.
SELECT * FROM Artist 
WHERE ArtistId NOT IN (SELECT ArtistId FROM Album)
```

```sql
-- 10. Listar todos los nombres de los artistas que comienza con la letra ’M’ y la cantidad de tracks, de esos artistas, con mas de 25 tracks, 
-- ordenado por cantidad de tracks de forma descendente

SELECT * FROM (SELECT a.Name, COUNT(*) as 'Canciones'
FROM Artist a
JOIN Album al on a.ArtistId = al.ArtistId
JOIN Track t on t.AlbumId = al.AlbumId
WHERE a.Name LIKE 'M%'
GROUP BY a.Name) AS Tablita
WHERE Canciones > 25

SELECT a.Name, COUNT(*) as 'Canciones'
FROM Artist a
JOIN Album al on a.ArtistId = al.ArtistId
JOIN Track t on t.AlbumId = al.AlbumId
WHERE a.Name LIKE 'M%'
GROUP BY a.Name
HAVING COUNT(*) > 25
```

```sql
-- 13. Para cada cliente obtener la cantidad de empleados que viven en la misma ciudad ordenados descendentemente por cantidad de empleados.
SELECT c.CustomerId, c.FirstName,c.LastName, COUNT(e.EmployeeId)
FROM Customer c
LEFT OUTER JOIN Employee e on e.City = c.City
GROUP BY c.CustomerId, c.FirstName, c.LastName
```

```sql
-- 14. Obtener los datos de todos los tracks del album ’Led Zeppelin I’. Hint: Utilizar consultas anidadas.
-- Sin consultas anidadas
SELECT *
FROM Track t, Album a
WHERE a.AlbumId = t.AlbumId AND a.Title = 'Led Zeppelin I'

-- Con consultas anidadas
SELECT * 
FROM Track
WHERE AlbumId in (SELECT AlbumId FROM Album WHERE Title = 'Led Zeppelin I')
```

```sql
-- 16. Obtener los playlist que no contengan ningun track de los albumes de los artistas “AC/DC” o “Audioslave” o ”Chris Cornell”
SELECT p.Name
FROM Playlist p
WHERE p.PlaylistId NOT IN (
    SELECT ptr.PlaylistId FROM PlaylistTrack ptr
    JOIN Track t on ptr.TrackId = t.TrackId
    JOIN Album al on al.AlbumId = t.AlbumId
    WHERE Al.ArtistId in (
        SELECT ArtistId from Artist 
        where Name in ('AC/DC', 'Audioslave', 'Chris Cornell')
    )
)

-- SOLO consultas anidadas
SELECT Name FROM Playlist WHERE PlaylistId NOT IN
    (SELECT PlaylistId FROM PlaylistTrack WHERE TrackId IN
        (SELECT TrackId FROM Track WHERE AlbumId IN 
            (SELECT AlbumId FROM Album WHERE ArtistId IN
                (SELECT ArtistId FROM Artist WHERE Name IN ('AC/DC', 'Audioslave', 'Chris Cornell')))))
```
