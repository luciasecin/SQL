```sql
-- Listar para cada track el nombre del mismo, el genero y el media type.
SELECT t.Name, g.Name as 'Genre', mt.Name as 'Media Type'
INTO #TempTableClasica
FROM Track t, Genre g, MediaType mt
WHERE t.GenreId = g.GenreId AND t.MediaTypeId = mt.MediaTypeId;

SELECT COUNT(*) AS 'Cantidad de Canciones' FROM #TempTableClasica;
SELECT * FROM #TempTableClasica;

-- Ahora lo hago con un join, deberian dar el mismo resultado
SELECT t.Name, g.Name as 'Genre', mt.Name as 'Media Type'
INTO #TempTableJoin
FROM Track t
    INNER JOIN Genre g ON t.GenreId = g.GenreId 
    INNER JOIN MediaType mt ON t.MediaTypeId = mt.MediaTypeId

-- Si no me da resultado es porque las tamblas son iguales
SELECT * FROM #TempTableClasica EXCEPT SELECT * FROM #TempTableJoin

DROP TABLE #TempTableClasica;
DROP TABLE #TempTableJoin;
```

```sql
-- Listar la cantidad de tracks que tiene cada genero.
-- Aca usamos left join porque puede que haya generos sin ningun track, si usaramos inner join y el genero no tiene tracks, no aparece en la lista
SELECT g.Name, COUNT(t.TrackId) as 'Cantidad de Tracks'
FROM Genre g
LEFT JOIN Track t on t.GenreId = g.GenreId
GROUP BY g.Name
```

```sql
-- Obtener los artistas que no tienen ´albumes. Dar al menos dos soluciones distintas.
SELECT art.Name AS 'Artistas sin albumes'
FROM Artist art
WHERE art.ArtistId NOT IN (SELECT ArtistId FROM Album)

SELECT art.Name AS 'Artistas sin albumes'
FROM Artist art 
LEFT OUTER JOIN Album alb on art.ArtistId = alb.ArtistId
GROUP BY art.Name
HAVING COUNT(alb.AlbumId) = 0

-- Esto era mucho mas facil que el count reina, los outer dan nulls!!! no olvidar
SELECT *
FROM Artist art
LEFT OUTER JOIN Album alb ON art.ArtistId = alb.ArtistId
WHERE alb.AlbumId IS NULL
```

```sql
-- Listar el nombre y la cantidad de tracks de los artistas con m´as de 50 tracks, ordenado por cantidad de tracks de forma ascendente.
SELECT art.Name, COUNT(t.TrackId) as cantidad
FROM Artist art
INNER JOIN Album alb on alb.ArtistId = art.ArtistId
INNER JOIN Track t on alb.AlbumId = t.AlbumId
-- Los artistas porian tener el mismo nombre asique hago group by tambien por el id
GROUP BY art.ArtistId, art.Name
HAVING COUNT(t.TrackId) > 50
ORDER BY cantidad ASC
```

```sql
-- Para cada cliente obtener la cantidad de empleados que viven en la misma ciudad ordenados descendentemente por cantidad de empleados.
SELECT c.CustomerId, COUNT(e.EmployeeId) AS 'Cantidad'
FROM Customer c
-- Left outer para que queden los demas customers en 0
LEFT OUTER JOIN Employee e ON c.City = e.City AND c.Country = e.Country AND c.[State] = e.[State] 
GROUP BY c.CustomerId
ORDER BY COUNT(e.EmployeeId) DESC
```
